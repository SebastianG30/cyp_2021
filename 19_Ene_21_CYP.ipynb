# Slicing (Continuación)

[ inicio: stop: incremento]

Notas:
1. El stop no es incluido.
2. El incremento se puede modificar.
3. Soporta indexado alternativo (negativos)

```
frase= "Como hacer a la gente mas positiva? muy facil, has que pierdan un electron!"
print(frase)
# imprimir "ace" de la plabra hacer
print(frase[6:9:1])


# Valores por defecto

print(frase[: 5: 2])

print(len(frase))
# Incremento
print(frase[: :2]) # Toda la frase
print(frase[ -1: : -1]) # Inverso -77

nombre= "Juan Jose"
print(nombre[1::-1])

primer=nombre[:4:]
print(primer)

print(nombre[-9::1])
```

# Métodos de la clase String(str)
En python todo es un objeto (P.O.O.), eso significa que cada tipo de dato trae consigo metodos (funciones) incluidos.

Para identificar los metodos disponibles se puede usar de dos funciones:
1. dir(str)  #Resumen de los metodos de esa clase.
2. help(str) # Información completa de los metodos de esa clase.

```
nombre= "Juan Jose"
print(nombre)
print( dir(str))

metodos=dir(str)
for metodo in metodos:
  print(metodo)
```

Los nombres de metodos que inician con doble gion bajo, significa que son metodos privados (POO), y no tendrias porque usarlos como programador consumidor.

Los demas metodos son publicos y son de libre uso.


```
print(nombre)
print(nombre.islower())
nombre="PEDRO FERNANDEZ"
print(nombre.capitalize())
print(nombre.endswith("Dez"))
print(nombre.replace("E", "**"))
ind_e=nombre.find("E")
print(ind_e)
nombre= nombre[ :ind_e:]+ "@" + nombre[ind_e+1::]
print(nombre)
print(nombre.replace("E","@",0))
print(nombre.lower())
print(nombre.upper())
nombre="josé josé, el principe de la canción"
print(nombre.count("josé"))
```

```
# Metodos para procesar Strings

numeros="  2  ,  1,  3 ,4,    5, 45,  7   "
print(numeros)
# Como extraigo los numeros de la cadena
# 1.- Eliminar los espacios al principio y al final con strip()
numeros=numeros.strip()
print(numeros)

print(help(str.split))
#2.- Separo por comas
numeros=numeros.split(",")
print(numeros)

#Paso 3, con un for e indices limpiar cadena de la lista
for indice in range(len(numeros) ):
  numeros[indice]= numeros[indice].strip()
print(numeros)
#Paso 4, convirtiendolos a tipo entero
for indice in range(len(numeros) ):
  numeros[indice]= int(numeros[indice])
print(numeros)

#Paso 5
```
